```python
from sklearn.preprocessing import LabelEncoder
lb = LabelEncoder()
```

```python
def encode_value(df,col_list):
    df1 = df.copy()
    for col in col_list:
        v1 = df1.groupby([col])['CO2 Emissions(g/km)'].mean().to_frame()
        v2 = v1['CO2 Emissions(g/km)'].to_dict()
        for i,j in df1.iterrows():
            comp = j[col]
            avg = v2[comp]
            df1.loc[i,col+'_encoded'] = avg
    return df1
```

```python
df2 = encode_value(df,['Make','Vehicle Class','Transmission','Fuel Type'])
```

### Multicollinearity
 It is caused when two or more independent varibales are correlated among themselves.

```python
cor_ = df[['Engine Size(L)','Cylinders','Fuel Consumption City (L/100 km)','Fuel Consumption Hwy (L/100 km)','Fuel Consumption Comb (L/100 km)','Fuel Consumption Comb (mpg)','CO2 Emissions(g/km)']].corr()
plt.figure(figsize = (10,10))
sns.heatmap(cor_,annot = True)
plt.show()
```

We can see a almost every variable has some correlations among themselves.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_x = df[['Engine Size(L)','Fuel Consumption City (L/100 km)',
       'Fuel Consumption Hwy (L/100 km)', 'Fuel Consumption Comb (L/100 km)',
       'Fuel Consumption Comb (mpg)']]
vif_data = pd.DataFrame()
vif_data['Feature'] = vif_x.columns
vif_data['VIF'] = [variance_inflation_factor(vif_x.values,i)for i in range(vif_x.shape[1])]
```

```python
vif_data
```

Variance inflation factor tells us how well the other predictor variables are able to explain the variance of a single predictor variable.Higher Inflation also provides us more evidence proving the existence of multicollinearity.
Fuel consumption combined has the highest vif. But we can try removing the fuel consumption of city and highway because combined fuel consumption gives us the fuel comsumption in both city and highways.

```python
from statsmodels.stats.outliers_influence import variance_inflation_factor
vif_x_1 = df[['Engine Size(L)', 'Fuel Consumption Comb (L/100 km)',
       'Fuel Consumption Comb (mpg)']]
vif_data1 = pd.DataFrame()
vif_data1['Feature'] = vif_x_1.columns
vif_data1['VIF'] = [variance_inflation_factor(vif_x_1.values,i)for i in range(vif_x_1.shape[1])]
```

```python
#Scaling of variables
def min_max(data):
    min_ = min(data)
    max_ = max(data)
    norm = [(x - min_)/(max_-min_) for x in data]
    return norm
```

```python
for i,j in x.iteritems():
#     print(i)
    a = min_max(j)
    x[i] = a
```

```python
y['CO2 Emissions(g/km)'] = min_max(y['CO2 Emissions(g/km)'])
```

### Homoscdasticity
 Homoscedasticity makes sure that the variance of the residuals which is the difference between the actual and the predicted values, is not affected by the predicted values or the predictors(independent variables)

```python
x
```

```python
from scipy import stats
import statsmodels.api as sm

#
X = sm.add_constant(x)
model = sm.OLS(y, X).fit()
#residuals is the difference between the observed and predicted values
residuals = model.resid

#predicted values
pred = model.predict(X)
sns.scatterplot(x = pred,y = residuals)
```

The variance of the residuals seems to vary as the predicted values change. Heteroscedasticity exists. Let us check which variable has caused the more effect.

```python
for i in x:
    sns.scatterplot(data = x , x = i , y = residuals)
    plt.title(f'Residual relationship with {i}')
    plt.show()
```

From the above scatterplots we can understand that the attributes fuel consumption in (l/km) and mpg has made impact on the variance of the residuals

### Absence of Auto correlation of errors

```python
x = df2[['Engine Size(L)','Fuel Consumption Comb (L/100 km)',
       'Fuel Consumption Comb (mpg)','Make_encoded',
       'Vehicle Class_encoded', 'Transmission_encoded', 'Fuel Type_encoded']]
y = df2[['CO2 Emissions(g/km)']]

X = sm.add_constant(x)
model = sm.OLS(y, X).fit()
#residuals is the difference between the observed and predicted values
residuals = model.resid

dwt = sm.stats.stattools.durbin_watson(residuals)
print(dwt)
```

The Durbin-Watson test value is not very high. Hence there is only a slight negative autocorrelation and will not affect the data significantly.

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error,r2_score
```

```python
y['CO2 Emissions(g/km)'] = min_max(y['CO2 Emissions(g/km)'])
# X = X.drop('const',axis = 1)
X['Engine Size(L)'] = min_max(X['Engine Size(L)'])
X['Fuel Consumption Comb (L/100 km)'] = min_max(X['Fuel Consumption Comb (L/100 km)'])
X['Fuel Consumption Comb (mpg)'] = min_max(X['Fuel Consumption Comb (mpg)'])
X['Make_encoded'] = min_max(X['Make_encoded'])
X['Vehicle Class_encoded'] = min_max(X['Vehicle Class_encoded'])
X['Transmission_encoded'] = min_max(X['Transmission_encoded'])
X['Fuel Type_encoded'] = min_max(X['Fuel Type_encoded'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_pred,y_test)
print("MSE",mse)
print("R2 Score",r2)
```

We have obtained good accuracy and a lower mean squared error. 

```python
X1 = X.drop(['Fuel Consumption Comb (L/100 km)'],axis = 1)
X_train, X_test, y_train, y_test = train_test_split(X1, y, test_size=0.2)

lr = LinearRegression()
lr.fit(X_train,y_train)
y_pred = lr.predict(X_test)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_pred,y_test)
print("MSE",mse)
print("R2 Score",r2)
```

The model provides an r2 score of 90.6 after removing the variable fuel consumption in litre per 100 kilo meter. The mean squared erroe also stays less around 0.0017. 

```python
from sklearn.preprocessing import PolynomialFeatures

for i in range(2,7):
    print(f"{i} degree")
    poly = PolynomialFeatures(i)
    X_train_poly = poly.fit_transform(X_train)
    X_test_poly = poly.transform(X_test)
    lr.fit(X_train_poly,y_train)
    y_pred = lr.predict(X_test_poly)
    mse = mean_squared_error(y_test,y_pred)
    r2 = r2_score(y_pred,y_test)
    print("MSE",mse)
    print("R2 Score",r2)
```

The polynomial regression model of degree 4 seems to provide a very better r2 score capable of explaining 99.4 percentage of the variance in the data.

```python
poly = PolynomialFeatures(4)
X_train_poly = poly.fit_transform(X_train)
X_test_poly = poly.transform(X_test)
lr.fit(X_train_poly,y_train)
y_pred = lr.predict(X_test_poly)
y_tr_pred = lr.predict(X_train_poly)
mse1 = mean_squared_error(y_train,y_tr_pred)
r2_ = r2_score(y_train,y_tr_pred)
mse = mean_squared_error(y_test,y_pred)
r2 = r2_score(y_pred,y_test)
print("Train MSE",mse1)
print("Train R2 Score",r2_)
print("Test MSE",mse)
print("Test R2 Score",r2)
```
